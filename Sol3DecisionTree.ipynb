# Importing Libraries and Reading data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("dark_background")
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import seaborn as sb
```

```python
data = pd.read_csv("C:/Users/Animkush/Desktop/Practice/data/irisrmc.csv")
```

```python
#data.head(5)
data = data.drop(["Id"], axis=1) # id column doesn't contribute anytihng to our data so we remove it.
data.head(5)
```

# Visualizing the data.

```python
iris = data
iris_g = sb.PairGrid(iris, hue="Species")
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
#creating labels and features.
X = data.drop(["Species"], axis=1)
y = data["Species"]
```

# Fitting the Data Features and Label into DecisionTreeClassifier

```python
dtree = DecisionTreeClassifier()
dtree.fit(X, y)
```

# Textual representation of Decision Tree after fitting the model. 

```python
text_rep= tree.export_text(dtree)
print(text_rep)
```

# Graphical Representation

```python
plt.rcParams.update({
    "text.color": "black"})
fig = plt.figure(figsize=(30,20))
_ = tree.plot_tree(dtree, feature_names=X.columns, filled=True, rounded=True)
```
